```python
#Capitalize
str = "i am dev patel."
solve = str.capitalize()
print(solve)
```

```python
#title
str = "i am dev patel"
solve = str.title()
print(solve)
```

```python
#Swapcase
str = "i am dev patel"
solve = str.swapcase()
print(solve)
```

```python
#CaseFold
str = "I AM DEV PATEL 123"
solve = str.casefold()
print(solve)
```

```python
#Center
str = "Hello"
solve = str.center(25)
print(solve)
```

```python
#Center with ! mark
str = "Hello"
solve = str.center(25,"!")
print(solve)
```

```python
#Count 
str = "Hello World Hello Hello hello "
solve = str.count("hello")
print(solve)
```

```python
#Endswith
str = "Hello World!"
solve = str.endswith('!')
print(solve)
```

```python
#Expand tabs
str = "H\te\tl\tl\to\t!"
solve = str.expandtabs()
print(solve)
```

```python
#index
str = "Hello everyone I am Dev Patel."
solve = str.index("Dev")
print(solve)
```

```python
#Checking string is alpha num
str = "hello1223"
solve = str.isalnum()
print(solve)
 
```

```python
#Checking string is alpha
str = "hello"
solve = str.isalpha()
print(solve)
```

```python
#Checking String is digit
str = "1234"
solve = str.isdigit()
print(solve)
```

```python
#Checking String is identifier
str = "Hello_World"
solve = str.isidentifier()
print(solve)
```

```python
#Checking String is lower
str = "Hello World"
solve = str.islower()
print(solve)
```

```python
#Checking String is upper
str = "HELLO WORLD"
solve = str.isupper()
print(solve)
```

```python
#Checking String is numeric
str = "93048395"
solve = str.isnumeric()
print(solve)
```

```python
#Checking String is space
str = " "
solve = str.isspace()
print(solve)
```

```python
#Checking String is title
str = "All The World Is Title"
solve = str.istitle()
print(solve)
```

```python
#Join the string 
str1 = ("Hello","World")
solve = " ".join(str1)
print(solve)
```

```python
#Lower
str = "I AM DEV PATEL"
solve = str.lower()
print(solve)
```

```python
#Upper case
str = "i am dev patel"
solve = str.upper()
print(solve)
```

```python
#Partition
str = "Hello, How are you?"
solve = str.partition("How")
print(solve)
```

```python
#Replace
str = "Hello World"
solve = str.replace("Hello","Hii")
print(solve)
```

```python
#Starts with
str = "Hello world"
solve = str.startswith("Hello")
print(solve)
```

```python
#Split
str = "Hello! How are you buddy?"
solve = str.split()
print(solve)
```
