```python
#20. Greatest number out of 3 using nested if else  
num1 = int(input("Enter First number: "))
num2 = int(input("Enter Second number: "))
num3 = int(input("Enter Third number: "))
if num1 > num2:
    if num1 > num3:
        print(num1," is the Greatest number")
    else:
        print(num3," is the Greatest number")
else:
    if num2 > num3:
        print(num2," is the Greatest number")
    else:
        print(num3," is the Greatest number")
```

```python
#30. find average of 5 number and give grade
list = []
for i in range(1,6):
    num = int(input(f"Enter Value of Number {i} out of 100.: "))
    list.append(num)
avg = (list[0] + list[1] + list[2] + list[3] + list[4])/5

if avg<50:
    print("Grade is F")
elif avg > 50 and avg < 60:
    print("Grade is D")
elif avg > 60 and avg < 70:
    print("Grade is C")
elif avg > 70 and avg < 80:
    print("Grade is B")
elif avg > 80 and avg < 90:
    print("Grade is A")
elif avg > 90:
    print("Grade is A+")
```
